## Feature Scaling (Standardization)

Before combining multiple features into a single score or feeding them into the model, it is essential to bring them onto a common scale.

---

### Why:

- Different features (e.g., funding amount, burn rate, growth intensity) operate on **very different ranges**
- Without scaling, features with larger magnitudes can **dominate model learning**
- Improves **model stability and convergence**
- Ensures **balanced contribution** of all features in:
  - Machine Learning model  
  - Composite metrics like *Startup Health Score*

---

### Approach:

We use **Standard Scaling (Z-score normalization)**, which transforms features to have:

- Mean = 0  
- Standard Deviation = 1  

$$
X_{scaled} = \frac{X - \mu}{\sigma}
$$

Where:  
- $\mu$ = mean of the feature  
- $\sigma$ = standard deviation  

---

### Features Scaled:

- funding_efficiency  
- funding_per_year  
- log_funding  
- burn_rate  
- runway  
- growth_intensity  

---

### Outcome:

- All features are **standardized to a comparable distribution**
- Prevents **bias from large-scale features**
- Improves **model generalization**
- Ensures **consistency between training and deployment (app.py)**

---

### Why Not Min-Max Scaling?

Min-Max scaling compresses values into a fixed range [0,1], but:

- It is **sensitive to outliers**
- Startup datasets (especially funding) are **highly skewed**
- It can distort feature importance in real-world distributions  

Hence, **StandardScaler is preferred** for better robustness and stability.

```python
from sklearn.preprocessing import StandardScaler

scale_features = [
    'funding_efficiency',
    'funding_per_year',
    'log_funding',
    'burn_rate',
    'runway',
    'growth_intensity'
]

scaler = StandardScaler()

X_train[scale_features] = scaler.fit_transform(X_train[scale_features])
X_test[scale_features]  = scaler.transform(X_test[scale_features])
```

## Final Data Cleaning Before Modeling

Before training machine learning models, we perform a final check to ensure that the dataset is free from missing and infinite values.

This step is important because some models (e.g., Logistic Regression) do not handle NaN or infinite values.

### Approach:
- Replace infinite values (caused by division operations) with NaN
- Fill all missing values with 0

This ensures a clean and stable dataset for model training.

```python
X = df[deploy_features].copy()

import numpy as np

# Clean invalid values
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(0, inplace=True)


print("Train NaN:", X_train.isna().sum().sum())
print("Test NaN:", X_test.isna().sum().sum())
```

##  Model Training

We train different models.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# models
lr = LogisticRegression(max_iter=1000, class_weight='balanced')

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced'
)

gb = GradientBoostingClassifier()

xgb = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    random_state=42,
    eval_metric='logloss'
)
cat = CatBoostClassifier(
    iterations=300,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=5,
    random_state=42,
    eval_metric='AUC',
    verbose=0
)
# train
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
gb.fit(X_train, y_train)
xgb.fit(X_train, y_train)
cat.fit(X_train, y_train)
```

## Deployment Model (Pipeline)

To ensure consistency between training and deployment, we create a pipeline using the most important features.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

from catboost import CatBoostClassifier
import joblib
import shap

deploy_features = [
    'funding_efficiency',
    'funding_per_year',
    'log_funding',
    'milestones',
    'relationships',
    'investment_rounds',
    'startup_age',
    'funding_rounds_count',
    'burn_rate',
    'runway',
    'growth_intensity',
    'health_score'
]

df_model['target'] = df['target']

X = df_model[deploy_features].copy()
y = df_model['target']

X.replace([np.inf, -np.inf], np.nan, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

scale = (y == 0).sum() / (y == 1).sum()

pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value=0)),
    ("scaler", StandardScaler()),
    ("model", CatBoostClassifier(
        iterations=300,
        learning_rate=0.05,
        depth=6,
        class_weights=[1, float(scale)],
        random_state=42,
        verbose=0
    ))
])

# TRAIN
pipeline.fit(X_train, y_train)

# PREDICT
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

# SAVE MODEL
joblib.dump(pipeline, "../models/pipeline.pkl")

# =========================
# SHAP
# =========================

# preprocess data (same as pipeline without model)
preprocessor = Pipeline(pipeline.steps[:-1])
X_train_transformed = preprocessor.fit_transform(X_train)

# train separate model for SHAP
shap_model = CatBoostClassifier(
    iterations=300,
    learning_rate=0.05,
    depth=6,
    class_weights=[1, float(scale)],
    random_state=42,
    verbose=0
)

shap_model.fit(X_train_transformed, y_train)

explainer = shap.TreeExplainer(shap_model)
shap_values = explainer.shap_values(X_train_transformed)

shap.summary_plot(shap_values, X_train_transformed, feature_names=deploy_features)

joblib.dump(explainer, "../models/shap_explainer.pkl")
```
